```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sig_f(value): # Sigmoid Function
    return 1  / (1 + np.exp(-value))
```

```python
dataset = pd.read_csv('training_project1.dat',
                      sep=' ', 
                      dtype=np.float32, 
                      names=['heel_sensor', 'toe_sensor'])

testset = pd.read_csv('recall_project1.dat',
                      sep=' ', 
                      dtype=np.float32, 
                      names=['heel_sensor', 'toe_sensor'])
```

```python
X = dataset.iloc[:,:].values
X_test = testset.iloc[:, :].values
```

```python
thd = 0.1
a = np.where(X[:,0] > thd, 1, 0) #IC
b = np.where(X[:,0] < thd, 1, 0) #HO
c = np.where(X[:,1] > thd, 1, 0) #FF
d = np.where(X[:,1] < thd, 1, 0) #TO
```

```python
def set(x):
    y = np.zeros(len(x))
    val = 1
    for i in range(1, len(x)):
        if x[i-1] <= 0 and x[i] >= val:
            y[i] = 1
        
    return y
```

```python
IC = set(IC)
HO = set(HO)
FF = set(FF)
TO = set(TO)
Mst = np.where(c-b <= 0, 0.0, c-b) + HO
swing = np.where(a+c >= 1, 1, 0)
```

```python
dataset['IC'] = IC.T
dataset['HO'] = HO.T
dataset['FF'] = FF.T
dataset['TO'] = TO.T
dataset['Mst'] = Mst.T
dataset['Swing'] = swing.T
dataset
```

```python
X_train = dataset.iloc[:, :2].values
Y_train = dataset.iloc[:, 2:].values
```

```python
from sklearn.utils import shuffle
X_train, Y_train = shuffle(X_train, Y_train)
```

```python
w_i_h = np.random.uniform(-1, 1, (12, 2))
w_h_o = np.random.uniform(-1, 1, (6, 12))
b_i_h = np.zeros((12, 1))
b_h_o = np.zeros((6, 1))
```

```python
u0 = 0.1 # initial learning rate
k = 0    
t = 1e9 # tao
```

```python
while True:
    se = 0
    for X, Y in zip(X_train, Y_train):
        X = X.reshape(2, 1)
        Y = Y.reshape(6, 1)

        # Layer 1 : input -> hidden
        h_pre = np.dot(w_i_h, X) + b_i_h
        h = sig_f(h_pre)
        # Layer 2 : hidden -> output
        o_pre = np.dot(w_h_o, h) + b_h_o
        o = sig_f(o_pre)
        
        # error
        se += (o - Y)**2
        
        # Learning rate
        u = u0 / (1 + (k/t))

        # Backpropagation : output -> hidden
        delta_o = (o - Y)
        w_h_o += -u * delta_o @ np.transpose(h)
        b_h_o += -u * delta_o
        
        # Backpropagation : hidden -> input
        delta_h = np.transpose(w_h_o) @ delta_o * (h * (1 - h))
        w_i_h += -u * delta_h @ np.transpose(X)
        b_i_h += -u * delta_h
        
    k += 1
    
    mse = se/len(X_train)
    print(np.average(mse))
    if np.average(mse) <= 1e-6:
        break
```

```python
cont = 0
y_pred = np.array([])
for i in X_test:
    # Layer 1 : input -> hidden
    h_pre = np.dot(w_i_h, i.reshape(2,1)) + b_i_h
    h = sig_f(h_pre)
    # Layer 2 : hidden -> output
    o_pre = np.dot(w_h_o, h) + b_h_o
    o = sig_f(o_pre)

    y_pred = np.append(y_pred, o)
    
y_pred = y_pred.reshape(len(X_test), 6)
y_pred[:, 0] = np.where(y_pred[:, 0] >= np.max(y_pred[:, 0])*0.99, 1, 0)
y_pred[:, 1] = np.where(y_pred[:, 1] >= np.max(y_pred[:, 1])*0.99, 1, 0)
y_pred[:, 2] = np.where(y_pred[:, 2] >= np.max(y_pred[:, 2])*0.99, 1, 0)
y_pred[:, 3] = np.where(y_pred[:, 3] >= np.max(y_pred[:, 3])*0.99, 1, 0)
y_pred[:, 4] = np.where(y_pred[:, 4] >= np.max(y_pred[:, 4])*0.99, 1, 0)
y_pred[:, 5] = np.where(y_pred[:, 5] >= np.max(y_pred[:, 5])*0.99, 1, 0)
```
